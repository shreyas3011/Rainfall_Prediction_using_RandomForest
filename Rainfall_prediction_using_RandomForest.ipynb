**data preprocessing**

```python
#drop highly corelated columns
data=data.drop(columns=['maxtemp','temparature','mintemp'])
```

```python
data.head()
```

```python
print(data.rainfall.value_counts())
```

```python
#seperate the majoraty and minority class
df_majority=data[data['rainfall']==1]
df_minority=data[data['rainfall']==0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
# downsample the majority class to match the count pf minority class

df_majority_downsampled=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_majority_downsampled,df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled.head()
```

```python
#shuffel the dataset
df_downsampled=df_downsampled.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
df_downsampled.head()
```

```python
#splitting the data into x and y
x=df_downsampled.drop(columns=['rainfall'])
y=df_downsampled['rainfall']
```

```python
x.head()
```

```python
y.head()
```

```python
#splitting the data into training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**model training**

```python
rf_model=RandomForestClassifier(random_state=42)

param_grid_rf={
    "n_estimators":[50,100,200],
    'max_features':['sqrt','log2'],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
grid_search_rf=GridSearchCV(estimator=rf_model,param_grid=param_grid_rf,cv=5,n_jobs=-1,verbose=2)
grid_search_rf.fit(x_train,y_train)
```

```python
best_rf_model=grid_search_rf.best_estimator_
print("best estimator for random forest :" ,grid_search_rf.best_params_)
```

**model evalution**

```python
cv_score=cross_val_score(best_rf_model,x_train,y_train,cv=5)
print("cross val score :",cv_score)
print("mean cross val score :",np.mean(cv_score))
```

```python
# test set performance
y_pred = best_rf_model.predict(x_test)

print("Test set Accuracy:", accuracy_score(y_test, y_pred))
print("Test set Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
